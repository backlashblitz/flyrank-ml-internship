## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import os
import pandas as pd


file_path = "data/raw/content_refresh_anonymized.csv"


if not os.path.exists(file_path):
    url = "https://raw.githubusercontent.com/backlashblitz/flyrank-ml-internship/main/data/raw/content_refresh_anonymized.csv"
    print("Downloading dataset from GitHub repository...")
    df = pd.read_csv(url)
else:
    df = pd.read_csv(file_path)


df["is_declining_label"] = df["trend_direction"].str.lower().eq("down").astype(int)


total_pages = len(df)
declining_rate = df["is_declining_label"].mean()


stale_pages = df[df["days_since_last_update"] >= 180]
stale_declining_rate = stale_pages["is_declining_label"].mean()


declining_impressions = df[df["is_declining_label"] == 1]["impressions_90d"].sum()


print(f"Number 1: Total pages = {total_pages:,} | Overall Declining Rate = {declining_rate:.1%}")
print(f"Number 2: Pages updated >= 180 days ago ({len(stale_pages):,}) have a declining rate of {stale_declining_rate:.1%}")
print(f"Number 3: Declining pages account for {declining_impressions:,} impressions over 90 days.")
```
